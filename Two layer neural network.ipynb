```python
#importing libraries needed
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import time
```

```python
#importing data
data=pd.read_csv('F:\coursera_,ml\machine-learning-ex3\ex3\\train.csv',sep=',')
```

```python
#splitting the traing data
x_train=data.iloc[:35000,1:]
y_train=data.iloc[:35000,0]
```

```python
x_train=x_train.as_matrix()
y_train=y_train.as_matrix()
```

```python
x_t=x_train.T/255               #flattening the features
y_train=y_train.reshape((len(y_train),1))
```

```python
y_t=np.zeros((len(y_train),10))
```

```python
for i in range(len(y_train)):
    for j in range(10):
        if j==y_train[i]:
            y_t[i,j]=1
```

```python
y_t=y_t.T
```

```python
#defining the sigmoid function
def sig(x):
    a=1/(1+np.exp(-1*x))
    return a
```

```python
# defining relu function
def relu(x):
    return x*(x>0)
```

```python
#defining relu derivative function
def relu_der(x):
    return 1*(x>0)
```

```python
#cost function
def cost(a,y):
        c=(-1/y.shape[1])*(np.sum(y*np.log(a)+(1-y)*np.log(1-a)))
        return c
```

```python
np.random.seed(1)
```

```python
alpha=0.00002     #setting the learning rate
c_history=np.zeros((2000,1))
m=len(y_t)
```

```python
#intializing the parameters randomly
w1=np.random.rand(25,784)*0.01
b1=np.random.rand(25,1)
w2=np.random.rand(10,25)*0.01
b2=np.random.rand(10,1)
```

Implementing the forward and backward propagation using the relu function as activation function

```python
for i in range(2000):
    tic=time.clock()
    
    #forward propagation
    z1=np.dot(w1,x_t)+b1
    a1=relu(z1)
    z2=np.dot(w2,a1)+b2
    a2=sig(z2)
    
    #computing cost
    j=cost(a2,y_t)
    c_history[i]=j
    
    #backward propagation
    dz2=a2-y_t
    dw2=(1/m)*np.dot(dz2,a1.T)
    db2=(1/m)*np.sum(dz2,axis=1,keepdims=True)
    dz1=np.dot(w2.T,dz2)*relu_der(z1)
    dw1=(1/m)*np.dot(dz1,x_t.T)
    db1=(1/m)*np.sum(dz1,axis=1,keepdims=True)
    
    #gradient descent
    w1=w1-alpha*dw1
    b1=b1-alpha*db1
    w2=w2-alpha*dw2
    b2=b2-alpha*db2
    
    toc=time.clock()-tic
    print('epoch:'+str(i)+' Time:'+str(toc)+' cost: '+str(j))
```

```python
xa=np.arange(1,2001,1)
plt.plot(xa,c_history)
plt.xlabel('iterations')
plt.ylabel('cost')
plt.show()
```

```python
pre_train=np.zeros((x_t.shape[1],1))
for i in range(x_t.shape[1]):
    pre_train[i]=np.argmax(a2[:,i])
```

```python
c=0
for i in range(5000):
    if pre_train[i]==y_train[i]:
        c=c+1
print(str(c/5000*100)+"%")
```

so the accuracy obtained on training dataset is around 93%
